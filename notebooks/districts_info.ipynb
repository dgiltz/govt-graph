```python
import os
from io import StringIO
import yaml
import json
import requests
from dotenv import load_dotenv

from pprint import pprint
```

```python
env_path = "../.env"
loaded = load_dotenv(env_path)
msg = "Succesfully loaded" if loaded else "Load failed"
msg
```

```python
datagov_api_key = os.environ["DATA_GOV_API_KEY"]

CKAN_API_ENDPOINT = "http://catalog.data.gov/api/3"
```

```python
output = requests.get(
    url=f"{CKAN_API_ENDPOINT}/action/package_search",
    params={"q": "congressional districts", "rows": 20},
    headers={"Authorization": f"Bearer {datagov_api_key}"}
)

json_out = json.loads(output.content.decode())

io = StringIO()
yaml.dump(json_out["result"], io, allow_unicode=True)
print(io.getvalue())
```
